```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("mtcars.csv")
print(data.to_string()) 
#mpg - Miles/(US) gallon
#cyl - Number of cylinders
#disp - Displacement, in cubic inches
#hp - Horsepower
#drat - Driveshaft ratio
#wt - Weight (1000 lbs)
#qsec - 1/4 mile time; a measure of acceleration
#vs - 'V' or straight - engine shape
#am - Transmission (0 = automatic, 1 = manual)
#gear - Number of forward gears
#carb - Number of carburetors
```

```python
data = data.drop(['disp', 'drat', 'carb'], axis=1)
data.head(5)
```

```python
data = data.rename(columns={"wt": "weight", "vs": "engine shape", "am": "transmission"})
data.head(5)
```

```python
duplicate_rows_df = data[data.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
data.carb.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of carburetors")
plt.ylabel('Number of cars')
plt.xlabel('carb');
```

```python
non_numeric_columns = ["model"]
numeric_data = data.drop(columns=non_numeric_columns)
correlation_matrix = numeric_data.corr()
```

```python
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
